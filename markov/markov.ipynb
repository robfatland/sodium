# Markov 

A Markov chain is a stochastic process named after Russian mathematician
[Andrey Markov](https://en.wikipedia.org/wiki/Andrey_Markov).


On the word 'state': A coin can be said to have two states,
heads and tails. The golden key chessboard problem is a good example of a 
non-trivial state insofar as there are 64 coins: Equivalent to a 64-bit
state. (It is useful to know offhand that $2^{10}$ is $1000$ and 
$2^20$ is one million. Apparently the chessboard has 16 million million 
million available states.)


The central idea of Markov processes is a system whose state in the near
future depends only on its state at $time \; = \; now$; there is no memory
of what has come before. 


We do get a refined view of what is meant by *now* (or *memor*) in Markov 
process theory from use of the word 'order'. A first order Markov chain 
is a sequence of states where the *next* state depends only on the *current* 
state. 'Second order' means the next state depends on the current and 
*previous* states; third order depends on the last *three* states; and so on. 
This stretches the definition of *now* to include a little bit of memory
of the recent past.


More terminology:
- random walk
- gambler's ruin
- martingale
- Stopped Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the parameter values for the Beta distribution
alpha_values = [0.5, 2.0, 0.5, 2.0]
beta_values = [0.5, 2.0, 2.0, 0.5]
x = np.linspace(0, 1, 100)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the Beta distributions
for i, (alpha, beta_param) in enumerate(zip(alpha_values, beta_values)):
    y = beta.pdf(x, alpha, beta_param)
    ax = axs[i // 2, i % 2]
    ax.plot(x, y, label=f'α={alpha}, β={beta_param}')
    ax.set_title(f'α={alpha}, β={beta_param}')
    ax.set_xlabel('x')
    ax.set_ylabel('Probability Density')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.suptitle('Shape of the Beta Probability Distribution', fontsize=16, y=1.02)
plt.show()
```

In contrast to the previous case: The AI is perfectly capable of reproducing
a well-established pdf.



## Learning prompt


The AI was prompted for an instructive graphical illustration of a Hidden Markov Model.


AI text: 
*To illustrate the concept of a Hidden Markov Model (HMM), you can create a simple visualization that shows the hidden states and the observable states, along with the transitions between them. Here's a Python code snippet using Matplotlib to create such a figure:*

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add hidden states
hidden_states = ['Sunny', 'Rainy']
G.add_nodes_from(hidden_states, layer='hidden')

# Add observable states
observable_states = ['Walk', 'Shop', 'Clean']
G.add_nodes_from(observable_states, layer='observable')

# Add edges with transition probabilities (example values)
edges = [
    ('Sunny', 'Sunny', 0.8), ('Sunny', 'Rainy', 0.2),
    ('Rainy', 'Sunny', 0.4), ('Rainy', 'Rainy', 0.6),
    ('Sunny', 'Walk', 0.6), ('Sunny', 'Shop', 0.3), ('Sunny', 'Clean', 0.1),
    ('Rainy', 'Walk', 0.1), ('Rainy', 'Shop', 0.4), ('Rainy', 'Clean', 0.5)
]

# Add edges to the graph
for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Position nodes using a layered layout
pos = {node: (0, i) for i, node in enumerate(hidden_states)}
pos.update({node: (1, i) for i, node in enumerate(observable_states)})

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10)

# Draw edge labels
edge_labels = {(u, v): f'{d["weight"]:.1f}' for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Hidden Markov Model: Hidden and Observable States')
plt.axis('off')
plt.show()
```
