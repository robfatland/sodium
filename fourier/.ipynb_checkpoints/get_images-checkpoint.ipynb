```python
import os, requests
from PIL import Image
import numpy as np

# unsplash: Free, high quality
os.makedirs('images', exist_ok=True)
image_urls = [
    'https://images.unsplash.com/photo-1506905925346-21bda4d32df4?w=800',  # Mountain landscape
    'https://images.unsplash.com/photo-1574169208507-84376144848b?w=800',  # Abstract architecture
    'https://images.unsplash.com/photo-1441974231531-c6227db76b6e?w=800',  # Forest/nature
    'https://images.unsplash.com/photo-1573164713714-d95e436ab8d6?w=800',  # Portrait/face
    'https://images.unsplash.com/photo-1518837695005-2083093ee35b?w=800',  # Ocean/water
    'https://images.unsplash.com/photo-1506905925346-21bda4d32df4?w=800',  # Urban/city
    'https://images.unsplash.com/photo-1574169208507-84376144848b?w=800',  # Animal/wildlife
    'https://images.unsplash.com/photo-1441974231531-c6227db76b6e?w=800',  # Texture/pattern
    'https://images.unsplash.com/photo-1573164713714-d95e436ab8d6?w=800',  # Food/objects
    'https://images.unsplash.com/photo-1518837695005-2083093ee35b?w=800'   # Sky/clouds
]

# Alternative: Use placeholder images with different patterns
placeholder_base = 'https://picsum.photos/800/800?random='

for i in range(10):
    try:
        # Use placeholder service for diverse random images
        url = f'{placeholder_base}{i+1}'
        
        print(f'Downloading image {i+1}/10...')
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        
        # Open image and convert to grayscale
        img = Image.open(requests.get(url, stream=True).raw)
        img_gray = img.convert('L')  # Convert to grayscale
        
        # Resize to 512x512
        img_resized = img_gray.resize((512, 512), Image.Resampling.LANCZOS)
        
        # Save as TIFF
        filename = f'images/image_{i+1:02d}.tiff'
        img_resized.save(filename, format='TIFF')
        
        print(f'Saved: {filename}')
        
    except Exception as e:
        print(f'Error downloading image {i+1}: {e}')
        
        # Create a synthetic grayscale pattern as fallback
        pattern = np.random.randint(0, 256, (512, 512), dtype=np.uint8)
        fallback_img = Image.fromarray(pattern, mode='L')
        filename = f'images/image_{i+1:02d}.tiff'
        fallback_img.save(filename, format='TIFF')
        print(f'Created fallback pattern: {filename}')

print('Done: See ./images/')
```
