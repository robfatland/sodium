# heat equation


We have a one-dimensional problem to solve concerning heat that leads down the garden path to Fourier analysis. 
One version of this problem invokes two metal rods at temperatures $a$ and $b$ that are put into contact end-to-end.
What is the temperature of this system as a function of position and time $T(x, t)$? Clearly we need the heat equation!


$\begin{align}
\frac{\partial}{\partial t}T(x,t) = \alpha \frac{\partial^2}{\partial x^2}T(x,t)
\end{align}$

## The grand trio


A further moment with Gil Strang: Pull the camera back for a moment (say in mathematical physics to 
choose a context): We have three grand differential equations to use as a foundation for future 
adventures: Laplace's equation (*elliptical*), the heat equation (*parabolic*) and the wave 
equation (*hyperbolic*).


Laplace we solve inside some closed (circular?) region. "The average of the surrounding region is zero." 
Steve Brunton connects this with the Poisson equation which is a non-zero Laplacian.


Heat and wave equations include time: You go forward in time. 


The heat equation is first order in time, matched to a second derivative in space with a
regulating constant $\alpha$ (or maybe $\alpha^2$ to emphasize it is positive).


The wave equation is second order in time, matching the second derivative in space with a
velocity coefficient $c^2$.

### Fourier analysis 1: Connecting the history to the challenging bit


We are told that Fourier made a tremendous jump while working on solving the heat equation. 
This jump decomposes arbitrary functions $f(x)$ into constituent sine and cosine functions.
Now for impatient people such as myself... well I have an intuition about the 
flow of heat, with the scars to prove it. Flowing heat is like overturning a bucket of motor oil 
on the center of the dining room table *blorp* and watching it spread out to make a big flat puddle. 
This in contrast with sines and cosines that bounce up and down. Why did Fourier need bouncy sine 
waves to deal with blorping flowing smearing heat?


Grant Sanderson has a narrative that completely demystifies the situation; and
I follow primarily the path he lays out in the next section. This section is for impatient
people; it is a little groundwork in three paragraphs. I want to first remark what a *solution*
actually is. First a solution has both a time and a space component:
The heat equation is a differential equation connecting a spatial distribution of an object's 
heat measured as *temperature* to the time evolution of temperature of that object considered
at a particular spot... like the center of the dining room table. Two views, then: A camera 
snapshot of temperature everywhere at one moment, and a tape recording of temperature over time 
at location $x$. Second, then, is an up-sell: We want those tape recordings of temperature with 
time at every possible location $x$ across the entire object. 


To re-state: The differential equation plus some *essential caveats* together pose the question 
"What happens to the distribution of heat in the object over time?" The solution Fourier sought 
amounted to "here is what happens... everywhere across the object for all time into the future". 
So why are sines and cosines part of the solution? Two reasons: First because sines and cosines
can be added together so as to accommodate those essential caveats I mentioned above. 
Second because even though the sines and cosines come into play at the first
instant of time $t = 0$, they will be released in time to the care and supervision of 
time-like oversight functions. Specifically the sines and cosines from time $0$ are to be 
squashed flat (like puddling motor oil) by *time decay functions*. Finally a generalization:
While these various sine and cosine functions can *individually* satisfy the rules imposed by the 
heat equation together with the essential caveats, they also *collectively* satisfy those rules 
when added together as a weighted sum. This is a property called *linearity*. And now: Back 
to a proper narrative. 

### Fourier analysis 2: The narrative


Following 3B1B and MIT OCW: We have an initial distribution of heat in one dimension (initial condition) and
a rule for location $x$ that says "the average of nearby temperature differenced with the 
temperature here at $x$ -- scaled by a heat flow regulating constant $\alpha$ -- drives the 
change in temperature here at $x$ over a small time step".


To be clear: "In one dimension" means a long thin (finite) bar of some uniform material.


More clearness: This problem includes no heat sources or sinks. These could be a blowtorch 
or radioactive decay (heat sources) or blackbody radiation (a sink, heat leaving the
bar). 


We can come up with a discrete solution that proceeds in time steps: Calculate the 
temperature along the bar at a moment in the future; update the bar; rinse and repeat.


However we need to regard the essential caveats. First is the distribution of temperature
at time $t = 0$. Second: Since the rod has two ends it has two boundary conditions 
governing how we plan to deal with those ends.


(I wonder how the heat equation works on a ring that has no boundary...)


Anyway: In sum there is the equation expressing the transmission of heat, the initial 
heat profile $T(x, 0)$ and the boundary condition rules. Deep breath. And now consider 
the initial heat profile happens to be a sine wave. It can be anything we like so why 
not a sine wave? Now the time-dependent solution can be expressed as the product
of temporal and spatial components, the latter being our sine wave.  


This with a little calculus earns us a separate heat component that moderates the spatial
temperature sine wave with time: A time-decay exponential, the decay owing to the sign
change on differentiating a sine wave twice. Now with both spatial and temporal components
of a solution this special sine wave case has the glow of "it satisfies the differential equation".


Next consider the boundary conditions that do not submit to the second derivative formalism
on the bar. Perhaps we deal with that using a "half" second derivative. Now the temperature 
at the ends of the bar exhibit zero temperature gradient... so while sine functions have 
worked to this point perhaps we will do better by phase shifting them over to cosines that
are flat at $x=0$. We can moderate the period of the cosine so it is also flat at the other
end of the bar at $x=L$. So now we have chosen an easy initial condition (cosine wave)
and matched it up to the bar so as to have zero derivative at the two ends. We have a
time component which is exponential decay, modified so as to properly satisfy the
differential equation $T_t = \alpha T_{xx}$.


Having dispensed with equation + initial + boundary we turn next to linearity: Many such
cosines will suffice, with a degree of freedom for each. It is left as an exercise for the
reader to show that any reasonable function can be approximated by stipulating an infinite
number of leading constant weights applied to each of the countable infinitude of cosines.
