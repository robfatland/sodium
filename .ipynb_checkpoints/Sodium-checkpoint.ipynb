## On the Recurring Theme of Distributed Information


In `Godel Escher Bach` the lead story is 'recursion leads 
to consciousness', a pretty decent punchline. Here I have no such punchline in
view. I am just interested in forms of *distributed information* as a loose theme.


Here to get started is an example: The dictionary has but one entry for `hippopotamus`.
This point-like, non-distributed alphabetized information suits our need for findability. 
There is the hippo, there is the dictionary entry. Language then steps up to serve the
role of information distribution: The hippo exists in a contextual web, a firmament of 
connected point-like definitions related to hippos: *mammals*, *rhinos*, *rivers* and so on. 


The Fourier transform in contrast jumps right into distributed information. It
replaces a set of individual atomic bits of information with a different set of atoms
in which the original set is still present but encoded. It is no longer recognizable. 
This is related to the eigenvector coordinate transform in linear algebra... where the 
story is something like "from a particular rule (linear operation) there is a different 
way of looking at this little world (new basis vectors) that makes life simpler 
(easier to solve problems)." 


Look at the mysteries surrounding us. Our history deciphering them so far
is riddled with the pattern of redistribution of information. Distributed
information seems to show up to aid us with discovery, with
understanding what is really going on. Table salt is table salt, and a star
is a star; what more do I need? Must I wonder what they are? The salt makes 
my food taste better, the twinkling light in the night sky is pretty; that
will suffice. And yet... I met a strange person today who tried to tell me 
there is sodium in the salt that can be found as well in the star. 


Maybe that's the punchline... to make progress we are obliged to let 
go our hold on established order and convenience.

### Notes on building out this ambitious agenda


* Gold Key Problem
    * Parity sets, XOR-no-carry: addition is the same as subtraction
        * ...and is multiplication then its own inverse?
    * There are lots of ways to encode a 6 bit number...
        * the very clever one yields the flip address from combining key and current 
        * specifically: The flip address = key address xor current address
    * chessboard address scheme is bitwise reading coin state as parity bit
        * tails is 1, heads 0
        * The scheme uses six parity bits to produce an address (2**6 = 64)
        * MSB to LSB: rank rank rank file file file
        * 1 1 1 1 1 1 is LRH corner, 1 1 1 0 0 0 is LLH corner
        * Each parity bit uses a unique mask of 32 squares
        * A mask is affected by its 32 squares, unaffected by the remaining 32
        * Mask * mask inner product: Use 1/-1 for in/out and sum the products
        * Define the board address as the compendium of six parity bits based on coin states
        * Exercise: Find all squares that do not affect the board address
        * Board configuration degeneracy: Many produce one parity address
            * There are 2^64 board configurations but only 2^6 parity addresses
        * key address xor initial parity address = flip address
        * Solution: A coin flip at F sets the parity address to the key address
    * We will see orthogonal functions in the context of basis functions (Fourier)
* metallicity
    * Kirkhoff and Frauenhofer
    * Kirkhoff's 3 laws of spectroscopy...
        * Hot, dense, any phase: Emits a continuous spectrum we call a blackbody spectrum. 
        * Low-density hot gas emits: Bright lines
        * Low-density cool gas between observer and blackbody absorbs: Absorption lines
    * Sodium flames
