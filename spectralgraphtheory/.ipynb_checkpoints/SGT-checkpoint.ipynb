# spectral graph theory



The idea here is to map a graph $G$ into some form of an adjacency matrix and then apply methods 
of linear algebra to look for insight into the structure of the graph.

```python
targetN = 40
targetE = 80
n = []
for i in range(1, targetN + 1):
    n.append((i, {"id": i}))

G = nx.Graph()
G.add_nodes_from(n)

e = []
E = 0
t = 0
while True:
    t = t + 1
    a = randint(1, targetN)
    b = randint(1, targetN)
    if not a == b and not (a, b) in e and not (b, a) in e:
        e.append((a, b))
        E += 1
    if E >= targetE: break

G.add_edges_from(e)
print('needed', t, 'tries to arrive at', E, 'edges')

ax = plt.subplot(111)
nx.draw(G, with_labels=True, font_weight='normal')
```
